**Problem Statement**: Given a list of integers, return True if the array contains a 3 next to a 3 somewhere.

Desired output:
has_33([1, 3, 3]) → True
has_33([1, 3, 1, 3]) → False
has_33([3, 1, 3]) → False
has_33([3,3,1]) → True



```python
def has_33(a):
  for index in range(len(a)-1):
    if(a[index] == 3):
      a[index + 1 ] == 3
      flag = True
    else:
      flag = False
  return flag
```

```python
a = [1,2,3,5,4,8,3,3]
```

```python
has_33(a)
```
